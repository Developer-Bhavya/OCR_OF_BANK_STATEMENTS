```python
import requests
from bs4 import BeautifulSoup
import os

# List of financial terms to search for images
search_terms = ["bank statement", "profit and loss statement", "balance sheet", "payslip", "passbook", "affidavit document"]

# Base URL for Bing image search
base_url = 'https://www.bing.com/images/search?q={}'

def download_images(search_term):
    # Encode the search term for use in the URL
    url = base_url.format(search_term.replace(' ', '+'))

    # Fetch HTML content from URL
    html = requests.get(url)
    content = html.text

    # Parse HTML content
    soup = BeautifulSoup(content, 'html.parser')

    # Find all <img> tags
    images = soup.find_all('img')

    # Create a directory for the current search term if it doesn't exist
    term_dir = f"downloaded_images/{search_term.replace(' ', '_')}"
    os.makedirs(term_dir, exist_ok=True)

    print(f"Found {len(images)} images for '{search_term}' on the page.")

    count = 0
    for i, image in enumerate(images):
        # Attempt to get the image URL
        image_url = image.get("data-srcset") or image.get("data-src") or image.get("data-fallback-src") or image.get("src")

        # Only proceed if an image URL is found
        if image_url:
            try:
                # Download the image
                img_data = requests.get(image_url).content
                # Save the image in the folder for the current search term
                with open(f"{term_dir}/image_{i+1}.jpg", "wb") as f:
                    f.write(img_data)
                count += 1
                print(f"Downloaded image {i+1} for '{search_term}'")
            except Exception as e:
                print(f"Could not download image {i+1} for '{search_term}': {e}")

    print(f"Downloaded {count} images out of {len(images)} found for '{search_term}'.")

# Loop over each search term and download images
for term in search_terms:
    download_images(term)

print("All images downloaded.")
```
